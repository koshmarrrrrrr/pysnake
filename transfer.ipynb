```python
import json
import pandas as pd

# Загрузить данные о вакансии из JSON-файла
with open('case_2_reference_without_resume_sorted.json', 'r', encoding='utf-8') as file:
    reference_data = json.load(file)

# Загрузить данные о члене команды из JSON-файла
with open('case_2_data_for_members.json', 'r', encoding='utf-8') as file:
    members_data = json.load(file)

# Извлечь информацию о вакансии
vacancy_uuid = reference_data['vacancy']['uuid']
vacancy_name = reference_data['vacancy']['name']
description = reference_data['vacancy']['description']

# Создать DataFrame для вакансии
vacancy_df = pd.DataFrame({
    'vacancy_uuid': [vacancy_uuid],
    'vacancy_name': [vacancy_name],
    'description': [description]
})

# Создать список для хранения ключевых навыков
key_skills_list = []

# Извлечь ключевые навыки из каждого резюме и добавить их в список
for member_data in members_data:
    key_skills = member_data['failed_resumes'][0]['key_skills']  # Предполагаем, что в списке failed_resumes есть только одно резюме
    key_skills_list.append(key_skills)

# Создать DataFrame для ключевых навыков
key_skills_df = pd.DataFrame({'key_skills': key_skills_list})

# Вывести первые строки обоих DataFrame для проверки
print(vacancy_df.head())
print(key_skills_df.head())
```

```python
import json

# Загрузить данные из JSON-файла
with open('case_2_data_for_members.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Создать список для хранения ключевых навыков
unique_skills = set()

# Итерация по резюме
for member in data:
    if 'failed_resumes' in member:
        resumes = member['failed_resumes']
        for resume in resumes:
            if 'key_skills' in resume:
                skills = resume['key_skills']
                if skills:
                    # Разделить ключевые навыки по запятым и добавить во множество
                    unique_skills.update(skills.split(', '))

# Вывести уникальные ключевые навыки
print("Уникальные ключевые навыки из всех резюме:")
for skill in unique_skills:
    print(skill)
```

```python
import json

# Загрузить данные из JSON-файла
with open('case_2_reference_without_resume_sorted.json', 'r', encoding='utf-8') as file:
    reference_data = json.load(file)

# Создать список для хранения данных резюме
resume_dataset = []

# Итерация по резюме
for resume_data in reference_data['resumes']:
    resume_info = {}
    resume_info['resume_uid'] = resume_data['uuid']
    resume_info['first_name'] = resume_data['first_name']
    resume_info['last_name'] = resume_data['last_name']
    resume_info['birth_date'] = resume_data['birth_date']
    resume_info['country'] = resume_data['country']
    resume_info['city'] = resume_data['city']
    resume_info['about'] = resume_data['about']
    resume_info['key_skills'] = resume_data.get('key_skills', "")
    resume_info['languages'] = resume_data.get('languageItems', [])
    
    # Обработка информации об образовании
    education_info = resume_data.get('educationItem', [])
    if education_info:
        education = education_info[0]  # Предполагаем, что образование одно
        resume_info['education_year'] = education.get('year', "")
        resume_info['education_organization'] = education.get('organization', "")
        resume_info['faculty'] = education.get('faculty', "")
        resume_info['specialty'] = education.get('specialty', "")
        resume_info['education_result'] = education.get('result', "")
        resume_info['education_type'] = education.get('education_type', "")
        resume_info['education_level'] = education.get('education_level', "")

    resume_dataset.append(resume_info)

# Вывести полученный датасет
for resume in resume_dataset:
    print(resume)
```

```python
import json

# Загрузить данные из JSON-файла
with open('case_2_reference_without_resume_sorted.json', 'r', encoding='utf-8') as file:
    reference_data = json.load(file)

# Создать список для хранения данных об опыте работы
experience_dataset = []

# Итерация по резюме
for resume_data in reference_data['resumes']:
    resume_uid = resume_data['uuid']
    experience_info = resume_data.get('experienceItem', [])
    
    # Итерация по опыту работы в резюме
    for experience in experience_info:
        experience_item = {}
        experience_item['resume_uid'] = resume_uid
        experience_item['starts'] = experience['starts']
        experience_item['ends'] = experience.get('ends', "")
        experience_item['employer'] = experience['employer']
        experience_item['city'] = experience['city']
        experience_item['position'] = experience['position']
        experience_item['description'] = experience['description']
        
        experience_dataset.append(experience_item)

# Вывести полученный датасет
for experience in experience_dataset:
    print(experience)
```

```python
combined_dataset = []

# Объединение данных из датасетов об опыте работы и данных о резюме
for resume_info, experience_info in zip(resume_dataset, experience_dataset):
    combined_resume = {}
    combined_resume.update(resume_info)  # Добавляем информацию о резюме
    combined_resume.update(experience_info)  # Добавляем информацию об опыте работы
    combined_dataset.append(combined_resume)

# Вывод полученного объединенного датасета
for data in combined_dataset:
    print(data)
```

```python
import json

# Load vacancy data from JSON file
with open('case_2_reference_without_resume_sorted.json', 'r', encoding='utf-8') as file:
    reference_data = json.load(file)

# Create a list to store vacancy data
vacancy_dataset = []

# Parse information about vacancies
for vacancy_data in reference_data:
    if vacancy_data == 'resumes':
        break  # Stop parsing when reaching 'resumes'
    
    vacancy_info = {}
    vacancy_info['uuid'] = reference_data[vacancy_data]['uuid']
    vacancy_info['name'] = reference_data[vacancy_data]['name']
    vacancy_info['keywords'] = reference_data[vacancy_data]['keywords']
    vacancy_info['description'] = reference_data[vacancy_data]['description']
    
    # Add vacancy information to the dataset
    vacancy_dataset.append(vacancy_info)

# Print the resulting dataset
for vacancy in vacancy_dataset:
    print(vacancy)
```

```python
import json

# Загрузить данные из JSON-файла
with open('case_2_data_for_members.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Создать список для хранения данных о резюме
resume_dataset = []

# Парсинг информации о резюме из отклоненных резюме
for item in data:
    failed_resumes = item.get('failed_resumes', [])
    confirmed_resumes = item.get('confirmed_resumes', [])
    
    for failed_resume in failed_resumes:
        resume_info = {}
        resume_info['status'] = 'failed'
        resume_info.update(failed_resume)  # Обновляем информацию о резюме
        resume_dataset.append(resume_info)
    
    for confirmed_resume in confirmed_resumes:
        resume_info = {}
        resume_info['status'] = 'confirmed'
        resume_info.update(confirmed_resume)  # Обновляем информацию о резюме
        resume_dataset.append(resume_info)

# Вывести полученный датасет
for resume in resume_dataset:
    print(resume)
```

```python
import json

# Загрузить данные из JSON-файла
with open('case_2_data_for_members.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Создать список для хранения данных о резюме
resume_dataset = []

# Парсинг информации о резюме из отклоненных резюме
for item in data:
    failed_resumes = item.get('failed_resumes', [])
    confirmed_resumes = item.get('confirmed_resumes', [])
    
    for failed_resume in failed_resumes:
        resume_info = {}
        resume_info['status'] = 'failed'
        resume_info.update(failed_resume)  # Обновляем информацию о резюме
        resume_dataset.append(resume_info)
    
    for confirmed_resume in confirmed_resumes:
        resume_info = {}
        resume_info['status'] = 'confirmed'
        resume_info.update(confirmed_resume)  # Обновляем информацию о резюме
        resume_dataset.append(resume_info)

# Записать полученный датасет в JSON-файл
with open('resume_dataset.json', 'w', encoding='utf-8') as outfile:
    json.dump(resume_dataset, outfile, ensure_ascii=False, indent=4)

print("Резюме успешно записаны в файл 'resume_dataset.json'")
```
